```python
df=df.drop(columns=['nameOrig'])
```

```python
df=pd.get_dummies(df,columns=['type'])
df
```

```python
df['d_type']=df['nameDest'].str[:1]
df
```

```python
df=df.drop(columns=['nameDest'])
```

```python
df=pd.get_dummies(df,columns=['d_type'])
df
```

```python
def norm(col):
  return (col-col.min())/(col.max()-col.min())
for i in ["step","amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]:
  df[i]=norm(df[i])
df
```

```python
from imblearn.over_sampling import SMOTE

# Assuming 'df' is your DataFrame
X = df.drop('isFraud', axis=1)
y = df['isFraud']

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
# Create a new DataFrame from the resampled data
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['isFraud'] = y_resampled
df_resampled
```
